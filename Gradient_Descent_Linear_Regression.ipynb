# Gradient Descent Linear Regression

This notebook demonstrates a simple linear regression model trained using Gradient Descent.

```python
# Import necessary libraries
import matplotlib.pyplot as plt
```

```python
# Data
x = [1, 2, 3, 4, 5]
y = [150, 200, 250, 300, 350]

# Initialize parameters
w = 50.0
b = 50.0

# Hyperparameters
lr = 0.01        # learning rate
epochs = 100    # number of iterations

n = len(x)
losses = []  # to store loss values
```

```python
# Training using Gradient Descent
for epoch in range(epochs):
    dw = 0
    db = 0
    total_loss = 0

    for i in range(n):
        f_x = w * x[i] + b            # prediction
        error = f_x - y[i]
        total_loss += error**2 / n    # MSE

        # Gradients
        dw += (2/n) * error * x[i]
        db += (2/n) * error

    # Update weights
    w -= lr * dw
    b -= lr * db

    losses.append(total_loss)

    # Print every 100 epochs
    if epoch % 100 == 0:
        print(f"Epoch {epoch}: Loss = {total_loss:.4f}, w = {w:.4f}, b = {b:.4f}")
```

## Final Model Parameters

```python
print(f"Final weight (w): {w}")
print(f"Final bias (b): {b}")
```

## Loss Curve

```python
plt.plot(losses)
plt.title("Loss over Epochs")
plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.grid(True)
plt.show()
```
